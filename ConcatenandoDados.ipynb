# **Concatenando dados**

```python
import numpy as np 
import pandas as pd 

from pandas import Series, DataFrame
```

```python
df = pd.DataFrame(np.arange(36).reshape(6,6))
df
```

```python
df2 = pd.DataFrame(np.arange(15).reshape(5,3))
df2
```

## **Concatedando dados**
**pd.concat([objeto_da_esqueda, objeto_da_direita])**

```python
pd.concat([df, df2])
```

```python
# forçar entrada dos dados nos indices 
pd.concat([df, df2], axis=1)
```
