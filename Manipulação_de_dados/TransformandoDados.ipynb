# **Transformando dados**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.arange(36).reshape(6,6))
df
```

# **Removendo Dados**
**nome_do_objeto.drop([índice de linhas])**

```python
df.drop([0, 2])
```

```python
# Removendo apenas colunas
df.drop([1,2], axis= 1)
```

# **Acresentando dados**
**DataFrame.join(objeto_da_esquerda,objeto_da_direita)**

```python
serie = pd.Series(np.arange(6))
serie.name = 'nova_variavel'
serie
```

```python
novo_df = pd.DataFrame.join(df, serie)
novo_df
```

# **Ordenando dados**

**nome_do_objeto.sort_values(by=[índice], ascending=[false])**

```python
df_ordenado = novo_df.sort_values(by=['nova_variavel'], ascending =False)
```
